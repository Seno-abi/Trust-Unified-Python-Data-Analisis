```python
#Import 
# Biasa
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

from numpy.random import randn

# Statistik
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bonus = sns.load_dataset('tips')
```

```python
bonus.head()
```

```python
sns.lmplot('total_bill','tip',bonus)
```

```python
sns.lmplot('total_bill','tip',bonus,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',bonus,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',bonus,fit_reg=False)
```

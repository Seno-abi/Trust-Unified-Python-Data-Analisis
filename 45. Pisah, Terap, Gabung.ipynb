```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df_bir = pd.read_csv('kualitas_bir.csv',sep=';')
df_bir.head()
```

```python
def ranker(df):
    df['alk_rank'] = np.arange(len(df)) +1
    return df
```

```python
df_bir.head()
```

```python
df_bir.sort_values('alkohol',ascending = False,inplace=True)
```

```python
df_bir = df_bir.groupby('kualitas').apply(ranker)
```

```python
nilai_kual =df_bir['kualitas'].value_counts()
```
